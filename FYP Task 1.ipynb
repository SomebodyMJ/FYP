```python
# Code 1 to find anything that is related to Conflict of Interest

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

# Path to your Chrome WebDriver executable
webdriver_path = r'C:\Users\Lee Ming Jia\Desktop\driver\chromedriver-win64\chromedriver.exe'

# Configure Chrome options
chrome_options = Options()
chrome_options.add_argument("--headless")  # Run Chrome in headless mode (without opening browser window)

# Create a new Chrome WebDriver
driver = webdriver.Chrome(service=Service(executable_path=webdriver_path), options=chrome_options)

# Get the URL
url = input("Please enter your Website: ")

# Navigate to the webpage
driver.get(url)

# Define the explicit wait with increased timeout
wait = WebDriverWait(driver, 20)

# Use XPath to find elements containing the text "conflict of interest" with case insensitivity and space normalization
xpath_expression = "//*[contains(translate(normalize-space(text()), 'ABCDEFGHIJKLMNOPQRSTUVWXYZ', 'abcdefghijklmnopqrstuvwxyz'), 'conflict of interest')]"

# Wait until elements containing the text "conflict of interest" are present using presence_of_all_elements_located
parent_elements_2 = wait.until(EC.presence_of_all_elements_located((By.XPATH, xpath_expression)))

print("Conflict of interest:")
print('')

for parent_element in parent_elements_2:
    # Find all following sibling elements of the located parent element
    following_siblings = parent_element.find_elements(By.XPATH, "./following-sibling::*")

    # Find all descendant elements of the located parent element
    descendants = parent_element.find_elements(By.XPATH, ".//descendant::*")

    # Combine the lists of following siblings and descendants
    combined_elements = following_siblings + descendants
    
    # Print the text of all combined elements
    for element in combined_elements:
        print(element.text)
        print('')
 
# Close the browser
driver.quit()
```

```python
# code 2 to find anything items that is related to Conflict of Interest in the tag or class name "ul"

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

# Path to your Chrome WebDriver executable
webdriver_path = r'C:\Users\Lee Ming Jia\Desktop\driver\chromedriver-win64\chromedriver.exe'

# Configure Chrome options
chrome_options = Options()
chrome_options.add_argument("--headless")  # Run Chrome in headless mode (without opening browser window)

# Create a new Chrome WebDriver
driver = webdriver.Chrome(service=Service(executable_path=webdriver_path), options=chrome_options)

# Get the URL
url = input("Please enter your Website: ")

# Navigate to the webpage
driver.get(url)

# Define the explicit wait with increased timeout
wait = WebDriverWait(driver, 20)

# Use XPath to find elements containing the text "conflict of interest" with case insensitivity and space normalization
xpath_expression = "//*[contains(translate(normalize-space(text()), 'ABCDEFGHIJKLMNOPQRSTUVWXYZ', 'abcdefghijklmnopqrstuvwxyz'), 'conflict of interest')]"

# Wait until elements containing the text "conflict of interest" are present using presence_of_all_elements_located
parent_elements = wait.until(EC.presence_of_all_elements_located((By.XPATH, xpath_expression)))

print("Conflict of interest:")
print('')

for parent_element in parent_elements:
    # Find all following sibling elements of the located parent element with tag 'ul' or class 'ul'
    following_siblings = parent_element.find_elements(By.XPATH, "./following-sibling::*[self::ul or contains(@class, 'ul')]")
    
    # Find all descendant elements of the located parent element with tag 'ul' or class 'ul'
    descendants = parent_element.find_elements(By.XPATH, ".//descendant::*[self::ul or contains(@class, 'ul')]")

    # Combine the lists of following siblings and descendants
    combined_elements = following_siblings + descendants
    
    # Print the text of all combined elements
    for element in combined_elements:
        print(element.text)
        print('')


# Close the browser
driver.quit()
```
